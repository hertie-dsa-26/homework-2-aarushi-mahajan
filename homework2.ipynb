- Accept a radius as input when instantiated.
- Have a method that calculates and returns the **perimeter** of the circle.
- Have a method that calculates and returns the **area** of the circle.

Import this class into the main Flask module so it can be used by your application.


````python
import math


class Circle:
    """A class to represent a circle."""

    def __init__(self, radius: float):
        """
        Initialise Circle with a radius.

        Parameters:
            radius (float): The radius of the circle.
        """
        self.radius = radius

    def perimeter(self) -> float:
        """
        Calculate and return the perimeter (circumference) of the circle.

        Returns:
            float: The perimeter of the circle.
        """
        return 2 * math.pi * self.radius

    def area(self) -> float:
        """
        Calculate and return the area of the circle.

        Returns:
            float: The area of the circle.
        """
        return math.pi * self.radius ** 2
````

## Question 2: Pointers and References (1 point)

Examine the following Python code carefully:

```python
def check_array(input):
    for idx in range(len(input)):
        if input[idx][0] == "1":
            input[idx] = None
    return input

original_array = ["1_3", "5_2"]
new_array = check_array(original_array)

print(original_array)

```

What does this code output? Explain **why** it produces that output, using the concepts of pointers and references in Python.

````python
# You may use this cell to experiment with the code above.
def check_array(input):
    for idx in range(len(input)):
        if input[idx][0] == "1":
            input[idx] = None
    return input

original_array = ["1_3", "5_2"]
new_array = check_array(original_array)

print(original_array)
print(new_array)
````

*Your answer here.*

Both arrays print the same result because Python passes lists by reference, not by copy. So when check_array() modifies the list, it changes the original object in memory. Both original_array and new_array point to the same list, so both reflect the change.

## Question 3: Algorithm Analysis (2 points)

Consider the following algorithm, which finds the contiguous subarray with the largest sum:

```python
max_sum = None
for i in range(len(array)):
    sum_subarray = 0
    for j in range(i, len(array)):
        sum_subarray += array[j]
        if max_sum is None or max_sum < sum_subarray:
            max_sum = sum_subarray
print(max_sum)
```

Analyze the efficiency of this algorithm:

1. Is this algorithm polynomial? Justify your answer.
2. What is the runtime of this algorithm in **Big-O** notation? Show your reasoning.

````python
import time

array = list(range(1000))

start = time.time()

max_sum = None
for i in range(len(array)):
    sum_subarray = 0
    for j in range(i, len(array)):
        sum_subarray += array[j]
        if max_sum is None or max_sum < sum_subarray:
            max_sum = sum_subarray

end = time.time()

print(max_sum)
print(f"Time taken: {end - start:.4f} seconds")
````

*Your answer here.*

This algorithm is polynomial. The runtime is expressed as a power of n (the input size), meaning it grows at a polynomial rate rather than exponentially. The runtime is O(n²) as there are two nested loops i.e., the outer loop runs n times, and for each iteration, the inner loop runs at most n times. This gives n × n = n² total operations.
